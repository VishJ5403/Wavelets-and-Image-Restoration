```python
def calculate_psnr(img1, img2):    #defining a function to calculate psnr
    # img1 and img2 have range [0, 255]
    img1 = img1.astype(np.float64)     #converting to float for ease of calculations
    img2 = img2.astype(np.float64)     #same
    mse = np.mean((img1 - img2)**2)    #taking mean squared error
    if mse == 0:
        return float('inf')      #case of mse zero i.e. infinite psnr
    return 20 * math.log10(255.0 / math.sqrt(mse))     #psnr being returned
```

```python
from math import log10, sqrt    #another try both will give same output
  
def PSNR(original, compressed):
    mse = np.mean((original - compressed) ** 2)
    if(mse == 0):  # MSE is zero means no noise is present in the signal .
                  # Therefore PSNR have no importance.
        return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse))
    return psnr
```

```python
def ssim(img1,img2,alpha=1,beta=1,gamma=1):     #defining ssim function with default alpha beta and gamma set as 1
    img1=img1.ravel()                 #creating a 1d array from 2d image array
    img2=img2.ravel()                 #same for 2nd image
#     X=np.stack((img1,img2), axis=0)
    cov1=np.cov(img1,img2)            #storing the covariance matrix
    sigma_xy=cov1[0][1]               #storing all different values 
    sigma_x2=cov1[0][0]               #of sigma xy, sigma x^2 and
    sigma_y2=cov1[1][1]               #sigma y^2 from cov matrix
    mu_x=np.mean(img1)                #calculating mean of data from img1 
    mu_y=np.mean(img2)                #calculating mean of data from img2
    l=(2*mu_x*mu_y+0.01)/((mu_x)**(2)+(mu_y)**(2)+0.01)         #calculating l(x,y)
    c=(2*sqrt(sigma_x2)*sqrt(sigma_y2)+0.01)/(sigma_x2+sigma_y2+0.01)        #calculating c(x,y)
    s=(sigma_xy+0.5)/(sqrt(sigma_x2)*sqrt(sigma_y2)+0.5)      #calculating s(x,y)
    ssim=(l**alpha)*(c**beta)*(s**gamma)                 #calculating ssim
    return ssim       #returning the ssim value
#i tried changing a few things nothing much i thought i had some error its i made a silly mistake of taking wrong values
# from the covariance matrix
```

```python
def gauss(ksize,sigma=1.0,mu=0.0):        #defining a function to create gaussian kernel
    x, y = np.meshgrid(np.linspace(-1,1,ksize), np.linspace(-1,1,ksize))        #making a grid of ksize x ksize
    d = np.sqrt(x*x+y*y)              
    g = np.exp(-( (d-mu)**2 / ( 2.0 * sigma**2 ) ) )          #storing the values calculated to make gaussian kernel in the grid
    return g
```

```python
g = gauss(5)
gauu_resu=signal.convolve2d(img3,g, mode='same')/np.sum(g)            #applying the gaussian kernel on target image
plt.imshow(gauu_resu,cmap='gray')        #result isnt visible to naked eye but
g                 
```

```python
test=cv2.imread('test.jpg')
work_image = cv2.cvtColor(test, cv2.COLOR_RGB2GRAY)

row,col= work_image.shape
mean = 0
var = 200
sigma = var**0.5
gauss = np.random.normal(mean,sigma,(row,col))
noisy = work_image + gauss

noisy[noisy>255]=255

noisy.max()              #doing the same process with diff image
```

```python
import numpy as np
import matplotlib.pyplot as plt

import pywt
import pywt.data


# Load image
original = noisy

# Wavelet transform of image, and plot approximation and details
titles = ['Approximation', ' Horizontal detail',
          'Vertical detail', 'Diagonal detail']
coeffs2 = pywt.dwt2(original, 'db2')              #taking waveletes
LL, (LH, HL, HH) = coeffs2                        #storing them
fig = plt.figure(figsize=(12, 3))
for i, a in enumerate([LL, LH, HL, HH]):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
    ax.set_title(titles[i], fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

fig.tight_layout()
plt.show()         
# Idea from :
#  https://pywavelets.readthedocs.io/en/latest/
```

```python
def rewavh(LH,HH,HL,T=1):        #defining a function to do hard thresholding with default threshold 1
    hlh=LH                       #storing LH in a diff array so that it doesnt affect the original array
    hhh=HH                       #storing HH in a diff array so that it doesnt affect the original array
    hhl=HL                       #storing HL in a diff array so that it doesnt affect the original array
    hlh[hlh<T] = 0               #setting all values below T to zero
    hhh[hhh<T] = 0
    hhl[hhl<T] = 0
    return pywt.idwt2((LL,(hlh, hhl, hhh)), 'db2')     #reconstructing the image from the updated wavelets
correctedh=rewavh(LH,HH,HL)
plt.imshow(correctedh,cmap='gray')
```

```python
PSNR(work_image, noisy)          #psnr of original and noisy image
```

```python
PSNR(work_image, correctedh)      #psnr of original and reconstructed image 
# as we can see psnr is higher hence image after reconstruction is better and hence follows the trend
```

```python
def rewavs(LH,HH,HL,T=1):     #defining a function to do soft thresholding
    slh=LH-T                  #as for soft thresholding we need to keep values greater than threshold value - threshold
    slh[slh<0] = 0            #and values less than threshold 0 lets store the wavelet - threshold value in an array and
    shh=HH-T                  #check if the values is less than 0 then keep it 0 otherwise as it is
    shh[shh<0] = 0 
    shl=HL-T
    shl[shl<0] = 0
    return pywt.idwt2((LL,(slh, shl, shh)), 'db2')        #reconstruction after soft thresholding
correcteds = rewavs(LH,HH,HL)
plt.imshow(correcteds,cmap='gray')
```

```python
PSNR(work_image, noisy)              #psnr of original and noisy image
```

```python
PSNR(work_image, correcteds)           #psnr of original and reconstructed image 
# as we can see psnr is higher hence image after reconstruction is better and hence follows the trend
# and psnr of soft thresholding is higher than hard thresholding 
```

```python
test=cv2.imread('test.jpg')
work_image = cv2.cvtColor(test, cv2.COLOR_RGB2GRAY)

row,col= work_image.shape
mean = 0
var = 200
sigma = var**0.5
gauss = np.random.normal(mean,sigma,(row,col))
noisy = work_image + gauss

noisy[noisy>255]=255

noisy.max()
```

```python
def denoising(img,ksize,var):                     #defining a denoising func
    s_ij2=0
    if(len(img.shape)==3):                        #checking for rgb or grayscaled image
        m,n,ch=img.shape
        for i in range(ch):
            for j in range(m):
                for k in range(n):
                    for a in range(int(k-ksize/2),int(k+ksize/2)):
                        for b in range(int(j-ksize/2),int(j+ksize/2)):
                            s_ij2+=img[b,a,i]**2
                    img[j,k,i]=1-(2*(var)*np.log(ksize**2)/s_ij2)               #updating the centre value of the patch of ksize
        return img
    else:
        m,n=img.shape
        for p in range(m):
            for q in range(n):
                for r in range(int(q-ksize/2),int(q+ksize/2)):
                    for s in range(int(p-ksize/2),int(p+ksize/2)):
                        s_ij2+=img[s, r]**2
                img[p,q]=1-(2*(var)*np.log(ksize**2)/s_ij2)                  #updating the centre value of the patch of ksize
        return img
```

```python
def denoisewav(img):               #defining a func to denoise all 3 wavelets and reconstruct an image
    coeffs2 = pywt.dwt2(img, 'db2')
    LL, (LH, HL, HH) = coeffs2
    dlh=denoising(LH,3,40000)
    dhl=denoising(HL,3,40000)
    dhh=denoising(HH,3,40000)
    return pywt.idwt2((LL,(dlh,dhl,dhh)), 'db2')
```

```python
final=denoisewav(noisy)
```

```python
PSNR(work_image, noisy)                   #psnr of original and noisy image
```

```python
PSNR(work_image, final)                   #psnr of original and reconstructed image 
# as we can see psnr is higher hence image after reconstruction is better and hence follows the trend
# and psnr of denoising is much better than soft thresholding as well as hard thresholding
```
