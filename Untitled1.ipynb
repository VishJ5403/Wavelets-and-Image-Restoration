```python
moire=cv2.imread("newspaper-dots.jpg",0)
plt.imshow(moire,cmap="gray")
a=moire.shape
rows=a[0]
cols=a[1]
print(a)
```

```python
padding=np.zeros((2*rows,2*cols),dtype=int) #padding to prevent aliasing("Prevention is better than cure!")
maskp2=padding
padding[:rows,:cols]=moire
plt.imshow(padding,cmap='gray')
```

```python
dft=cv2.dft(np.float32(padding),flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
findft=dft_shift
magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))  
plt.subplot(121),plt.imshow(padding, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum')
plt.show()
#plotting magnitude spectrum by taking dft shifting it using fft shift and taking magnitude of it
```

```python
centre=(320,477)
radius=60
initmask=np.where(magnitude_spectrum>230,0,255)
# initmask=cv2.circle(b,centre,radius,(0,0,0),-1)
b=initmask
# plt.imshow(initmask)
# initmask-=255
plt.imshow(initmask,cmap='gray')
# creating a mask of my own using thresholding and based on what portion of image i want
```

```python
initmask=cv2.circle(b,centre,radius,255,-1)
plt.imshow(initmask,cmap='gray')
initmask.shape
#same as above
```

```python
initmask[354:610,317:323]=255
initmask[467:487,188:452]=255
plt.imshow(initmask,cmap='gray')
#final part of the mask
```

```python
initmask//=255
initmask
#making values in the mask 0 or 1 so as to not scale values after multiplying with dft
```

```python
dft_shift[:,:,0]*=initmask
dft_shift[:,:,1]*=initmask
findft=dft_shift
magnidft = 20*np.log(1+cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))
plt.imshow(magnidft,cmap='gray')
```

```python
idft_shift=np.fft.ifftshift(findft)
i_dft=cv2.idft(idft_shift)
img_back = cv2.magnitude(i_dft[:,:,0],i_dft[:,:,1])
finimg=img_back[:477,:320]
finimg=finimg**0.55
plt.subplot(121),plt.imshow(finimg, cmap = 'gray')
plt.title('Final'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(moire, cmap = 'gray')
plt.title('Moire'), plt.xticks([]), plt.yticks([])
plt.show()
#final image after inverse dft
#i actually removed my other tries,didnt read the instructions well but i tried gaussian blur but there was too 
# much loss of information and it wasnt as helpful and i tried a few filters but in some moire was still there or
# it was degrading the image a lot
```
